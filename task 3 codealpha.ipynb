```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
import zipfile
import pandas as pd
import io

# Assuming the uploaded file is 'archive (6).zip' and contains 'Student_Marks.csv'
# Find the name of the uploaded zip file
zip_file_name = list(uploaded.keys())[0]

# Unzip the file in memory
with zipfile.ZipFile(io.BytesIO(uploaded[zip_file_name]), 'r') as z:
    # Assuming 'Student_Marks.csv' is inside the zip file
    with z.open('Student_Marks.csv') as f:
        data = pd.read_csv(f)

print(data.columns)
```

```python
print(data.columns)
```

```python
plt.hist(data["Marks"])

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Distribution of Marks")

plt.show()
```

```python
plt.scatter(data["time_study"], data["Marks"])

plt.xlabel("Study Time")
plt.ylabel("Marks")
plt.title("Study Time vs Marks")

plt.show()
```

```python
data["number_courses"].value_counts().plot(kind='bar')

plt.xlabel("Number of Courses")
plt.ylabel("Count")
plt.title("Courses Distribution")

plt.show()
```

```python
sns.boxplot(data["Marks"])

plt.title("Box Plot of Marks")

plt.show()
```

```python
sns.heatmap(data.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.plot(data["time_study"], data["Marks"])

plt.xlabel("Study Time")
plt.ylabel("Marks")
plt.title("Study Time vs Marks Line Plot")

plt.show()
```
